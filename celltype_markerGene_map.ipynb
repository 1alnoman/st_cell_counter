```python
markers = dc.get_resource('PanglaoDB')
```

```python
markers_filtered = markers[markers['human'] & markers['canonical_marker'] & ((markers['human_sensitivity'] >= 0.5) | (markers['human_specificity'] >= 0.1))]
```
